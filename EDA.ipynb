```python
import os
import matplotlib.pyplot as plt
```

```python
dataset = {'helmet':0, 'person':1, 'fire':2, 'safetyvest':3}

helmet_count = 0
person_count = 0
fire_count = 0
safetyvest_count = 0
```

```python
path_to_images = './YOLO_dataset/img/'

labels = []

# Go through each file in the folder
for file_ in os.listdir(path_to_images):
    # if the file ends with 'txt' open
    if file_[-3:] == 'txt':
        txt_file = open(path_to_images + file_, 'r')
        # read the lines in the file and append the first character (label) to labels list
        lines = txt_file.readlines()
        for line in lines:
            stripped = line.strip()
            labels.append(stripped[0])
    
print(len(labels))
```

```python
for label in labels:
    if label == '0':
        helmet_count += 1
    if label == '1':
        person_count += 1
    if label == '2':
        fire_count += 1
    if label == '3':
        safetyvest_count += 1
        
print(helmet_count, person_count, fire_count, safetyvest_count)
print(len(labels))
print(helmet_count + person_count + fire_count + safetyvest_count)
```
